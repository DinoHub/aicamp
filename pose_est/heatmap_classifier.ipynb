```python
from matplotlib import pyplot as plt
from matplotlib import image as mim
%matplotlib notebook
```

```python
train_set = []
for clsname in os.listdir(til_data_path):
    class_dir = os.path.join(til_data_path, clsname)
    for imgname in os.listdir(class_dir):
        clsname = clsname.lower()
        pid = imgname.split('.')[0].split('_')[-1]
        train_set.append((clsname, pid)) 
```

```python
highkneel_dir = os.path.join(til_data_path, 'HighKneel')
for imgname in os.listdir(highkneel_dir):
    if '_3.png' in imgname:
        plt.figure()
        print(imgname)
        img = mim.imread(os.path.join(highkneel_dir, imgname))
        plt.imshow(img)
```

```python
train_count = 0
test_count = 0
popping = deepcopy(train_set)
train_npy = []
test_npy = []
for npy_name in npy_names:
    if npy_name.startswith('LeopardCrawl'):
        print(npy_name)
        continue
    split = npy_name.split('_')
    if len(split) == 3:
        classname, pid, uid = split
        classname = classname.lower()
        if (classname, pid) in train_set:
            train_count+= 1
            train_npy.append(npy_name)
            popping.remove((classname, pid))
        else:
            test_count += 1
            test_npy.append(npy_name)
    elif len(split) == 2:
        print(npy_name)
        
#     if pid == '3' and classname == 'HighKneel':
#         print(npy_name)
#         imggg = np.load(os.path.join(data_path, npy_name))
#         img = imggg[:,:,:3]
#         print(img.shape)
#         plt.figure()
#         plt.imshow(img)

print('remain:', popping)
```
