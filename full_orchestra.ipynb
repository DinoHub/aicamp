```python
def get_labels(test_folder):
    class_labels = sorted( list(os.listdir(test_folder)) )
    labels = []
    for cl in class_labels:
#         print('--> class: {}'.format(cl))
        pose_folder = os.path.join( test_folder, cl )
        sorted_imgs = sorted(list(os.listdir( pose_folder )))
        for im in sorted_imgs:
            img_path = os.path.join( pose_folder, im )
            labels.append((img_path,cl))
    import pickle
    with open('labels.p', 'wb') as fnp:
        pickle.dump(labels, fnp )
    return labels
```

```python
npy_dir = '/home/dh/Workspace/aicamp/predictions/full_train'
list_of_softs = []
for npy_name in os.listdir(npy_dir):
    npy_path = os.path.join(npy_dir, npy_name)
    list_of_softs.append(np.load(npy_path))
```

```python
np.array(list_of_softs).shape
X = np.zeros((2343, 150))
for i, softs in enumerate(list_of_softs):
    for j, soft in enumerate(softs):
#         print(i, j)
#         print(data.shape)
        X[j, i*15:(i+1)*15] = soft
```

```python
labels_class = [x[1] for x in labels]
labels_class
```

```python
sorted_ = list(set(labels_class))
sorted_.sort()
label2idx = {cn:idx for idx, cn in enumerate(sorted_)}
label2idx
```

```python
y = [label2idx[x] for x in labels_class]
```

```python
from keras.utils import to_categorical
y_1hot = to_categorical(y, num_classes = 15) 
```

```python
y_1hot.shape
```

```python
for x in X:
    vec = x[0:15]
    print(sum(vec))
```

```python
X.shape
reshapes = []
for x in X:
    intermediate = x.reshape((10, -1))
#     intermediate = x.reshape((15, -1)).T
    reshapes.append(intermediate)
reshapes = np.array(reshapes)
reshapes.shape
```

```python
X_reshaped = reshapes
X_reshaped = np.expand_dims(X_reshaped, axis=-1)
X_reshaped.shape
```

```python
for x in X_reshaped:
    for xx in x:
        print(sum(xx))
```

```python
from keras.layers import Conv2D, Flatten, Input, Dense, GlobalAveragePooling2D, BatchNormalization, Activation
from keras.models import Model, load_model 
```

```python
# build MLP
input_shape = (10,15,1)
input_features = Input(shape=input_shape)
# x = Conv2D(1, (10,1), strides = 1, activation='relu', bias_initializer='zeros', kernel_initializer='zeros')(input_features)
x = Conv2D(1, (10,1), strides = 1, activation='relu', bias_initializer='zeros', kernel_initializer='ones')(input_features)
x = Flatten()(x)
predictions = Activation('softmax')(x)
orchestra_model = Model(inputs=input_features, outputs = predictions)
```

```python
orchestra_model.summary()
```

```python
from keras.optimizers import Adam
```

```python
optz = Adam(lr=0.001)
orchestra_model.compile(optimizer=optz,
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
bs = 64
n_epoch = 5000
context = 'full_orchestra1'
```

```python
import shutil
if not os.path.isdir('models/{}'.format(context)):
    os.makedirs('models/{}'.format(context))
if os.path.isdir('./tblogs/{}_tblogs'.format(context)):
    shutil.rmtree('./tblogs/{}_tblogs'.format(context))
```

```python
from keras.callbacks import ModelCheckpoint, CSVLogger, TensorBoard, LearningRateScheduler
csvLogger = CSVLogger('logs/{}.log'.format(context))
valLossCP = ModelCheckpoint('models/{}/{}_loss.hdf5'.format(context, context), save_best_only=True)
valAccCP = ModelCheckpoint('models/{}/{}_acc.hdf5'.format(context, context), monitor='val_acc', save_best_only=True)
tbCallback = TensorBoard( log_dir='./tblogs/{}_tblogs'.format(context), histogram_freq=0, write_graph=True, write_images=True, update_freq='epoch' )
```

```python
all_callbacks = [csvLogger, valLossCP, valAccCP, tbCallback]
```

```python
orchestra_model.fit(x=X_reshaped, y=y_1hot, batch_size=bs, epochs=n_epoch, validation_split=0.15, callbacks=all_callbacks)
```
